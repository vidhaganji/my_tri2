2) example for burglar alarm

```python
def NOT_gate(A):
    return not A
def NAND_gate(A, B):
    return not (A and B)
def NOR_gate(A, B):
    return not (A or B)
def XOR_gate(A, B):
    return (A and not B) or (not A and B)
def XNOR_gate(A, B):
    return (A and B) or (not A and not B)
def burglar_alarm_circuit(A, B, C):
    return AND_gate(OR_gate(A, B), NOT_gate(C))
print('A', 'B', 'C', "-->", 'Output')
inputs = [(A, B, C) for A in [False, True] for B in [False, True] for C in [False, True]]
for A, B, C in inputs:
    print(A, B, C, "-->", burglar_alarm_circuit(A, B, C))

```

**reasonable time** = means number of steps the algorithm takes is less than or equal to a polynomial function

```python
import timeit
import random

def scale_list(original_list, factor):
    scaled_list = original_list * factor
    return scaled_list
original_list = [random.randint(0, 100) for _ in range(100)]
scaling_factor = 10
time_taken = timeit.timeit(lambda: scale_list(original_list, scaling_factor), number=10000)
print("Time taken to scale list by a factor of {}: {} seconds".format(scaling_factor, time_taken))

```

```python
import time
def fibonacci_recursive(n):
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)
def fibonacci_iterative(n):
    if n <= 1:
        return n
    else:
        a, b = 0, 1
        for _ in range(n - 1):
            a, b = b, a + b
        return b
def measure_time(func, arg):
    start_time = time.time()
    result = func(arg)
    end_time = time.time()
    return result, end_time - start_time
small_n = 10
fib_recursive_small, time_recursive_small = measure_time(fibonacci_recursive, small_n)
fib_iterative_small, time_iterative_small = measure_time(fibonacci_iterative, small_n)
large_n = 35
fib_recursive_large, time_recursive_large = measure_time(fibonacci_recursive, large_n)
fib_iterative_large, time_iterative_large = measure_time(fibonacci_iterative, large_n)

```

code for multiplication of negative numbers:
```python
def multiply(x, y):
    if x < 0 and y < 0:
        return multiply(abs(x), abs(y))
    elif x < 0 or y < 0:
        return -multiply(abs(x), abs(y))
    else:
        result = 0
        for _ in range(y):
            result += x
        return result
```
